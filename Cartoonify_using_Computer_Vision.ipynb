```python
#Importing required libraries
import cv2 # OpenCV is used for image processing and machine learning software library
from google.colab.patches import cv2_imshow # cv2_imshow() method is used to display an image in a window
from skimage import io # is ysed to read,save and display image or video
```

```python
#Reading image 
img = cv2.imread("/content/drive/MyDrive/2013641_miniproject( Cartoonify using Machine Learning )/balckie.jpg")
```

```python
# Displaying image using cv2 library
cv2_imshow(img)
```

```python
#displaying img using io libraries
io.imshow(img)
```

```python
#Converting to RGB color
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
io.imshow(img)
```

```python
# Converting image to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 5)
io.imshow(gray)
```

```python
# ADAPTIVE THRESHOLDING TECHNIQUE
edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
io.imshow(edges)
```

```python
# SIMPLE THRESHOLDING TECHNIQUE
edges1 = cv2.threshold(gray,127, 255, 0)[1]
io.imshow(edges1)
```

```python
#Cartoonifying the image
color = cv2.bilateralFilter(img, 9, 250, 250)
io.imshow(color)
```

```python
# CARTOON USING ADAPTIVE THRESHHOLD TECHNIQUE
cartoon = cv2.bitwise_and(color, color, mask=edges)
io.imshow(cartoon)
```

```python
# CARTOON USING SIMPLE THRESHHOLD TECHNIQUE
cartoon1 = cv2.bitwise_and(color, color, mask=edges1)
io.imshow(cartoon1)
```
